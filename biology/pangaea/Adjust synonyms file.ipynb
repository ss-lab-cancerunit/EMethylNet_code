# Adjust synonyms file

The provided synonyms file contains some English words that are falsely identified as genes. Eg, skin can be a gene synonym. This notebook removes these words from the synonyms file to avoid this problem.

This may result in missing some genes that have English word synonyms.

```python
# read in synonyms file
import json
synonyms = json.load(open('gene_to_synonyms.txt')) 
```

```python
from nltk.corpus import words
eng_words = set(words.words())
```

```python
trimmed_synonyms = {}
```

```python
for gene in synonyms:
    new_syns = [s for s in synonyms[gene] if s not in eng_words ]
    trimmed_synonyms[gene] = new_syns
```

```python
for i in trimmed_synonyms:
    if len(trimmed_synonyms[i]) != len(synonyms[i]):
        print(len(trimmed_synonyms[i]), len(synonyms[i]))
        # checking the lengths have decreased, to check its working
```

```python
json.dump(trimmed_synonyms, open('genes_to_synonyms_trimmed.txt', 'w'))
```
