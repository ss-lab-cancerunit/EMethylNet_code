# Plot feature stats

Reading in the feature stats for every cancer type and plotting them all on one plot.

```python
stringency = 'relaxed'
cancer_types = ['BLCA', 'BRCA', 'COAD', 'ESCA', 'HNSC', 'KIRC', 'KIRP', 'LIHC', 'LUAD', 'LUSC', 'PRAD', 'THCA', 'UCEC']
```

```python
# read in all stats:
import pandas as pd
stats = [pd.read_csv('stats_min_count_1__' + cancer_type + '_' + stringency + '.csv', sep = ' ') for cancer_type in cancer_types]
```

```python
# we just want the totals:
totals = [stats[i].loc['totals'] for i in range(len(stats))]
```

```python
# read in multiclass probes and plot
cancer_type = ''
import pandas as pd
stats_multi = pd.read_csv('stats_min_count_1__' + cancer_type + '_' + stringency + '.csv', sep = ' ')

# we just want the totals:
totals_multi = stats_multi.loc['totals']

totals.append(totals_multi)
```

```python
# turn this into a data frame:
totals = pd.DataFrame(totals)
totals.index = cancer_types + ['Multiclass']
totals = totals.transpose()
```

```python
# make a scatter plot:

import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
palette = sb.color_palette('Paired')
palette.append((0.43,0.43,0.43))
palette.append((0,0,0)) # adding on two more colours to get 14 colours

for i in range(len(cancer_types)):
    ax = sb.scatterplot(x = range(9), y = cancer_types[i], data = totals, color = palette[i], label = cancer_types[i], s = 150)
#     ax = sb.lineplot(x = range(9), y = cancer_types[i], data = totals, color = palette[i], label = cancer_types[i])
    

plt.xticks(range(9), list(totals.index), rotation = 20)
plt.ylabel('Probe frequency')
plt.legend(loc = 'right', bbox_to_anchor = (0.7, 0.5, 0.6, 0))
```

```python
totals = totals.transpose()
```

```python
# Note: take off the multiclass when looking at this plot, it is not very clear with multiclass added
tidy = totals.stack()

tidy = tidy.reset_index()
tidy.columns = ['Cancer type', 'x', 'y']
# pd.DataFrame(tidy)
plt.figure(figsize=(8,6))
ax = sb.swarmplot(data = tidy, x = 'x', y = 'y', hue = 'Cancer type', palette = palette, s = 7)

plt.xticks(rotation = 20)
plt.ylabel('Probe frequency')
plt.legend(loc = 'right', bbox_to_anchor = (0.7, 0.5, 0.6, 0))
```

```python
# now I want to turn the absolute numbers into ratios to compare the proportions
totals_ratio = totals.divide(totals['genes_mapped'], axis = 'index')
# totals_ratio


# add open sea column (ie not near an island)
open_sea = 1 - (totals_ratio['N_Shelf'] + totals_ratio['N_Shore'] + totals_ratio['Island'] + totals_ratio['S_Shore'] + totals_ratio['S_Shelf'])
totals_ratio['Open_sea'] = open_sea

# swap open sea and genes mapped
new_cols = list(totals_ratio.columns)
temp = new_cols[-2]
new_cols[-2] = new_cols[-1]
new_cols[-1] = temp
totals_ratio = totals_ratio[new_cols]

tidy = totals_ratio.stack()

tidy = tidy.reset_index()
tidy.columns = ['Cancer type', 'x', 'y']
# pd.DataFrame(tidy)
plt.figure(figsize=(8,6))
ax = sb.swarmplot(data = tidy, x = 'x', y = 'y', hue = 'Cancer type', palette = palette, s = 7)

plt.xticks(rotation = 20)
plt.ylabel('Probe frequency')
plt.legend(loc = 'right', bbox_to_anchor = (0.7, 0.5, 0.6, 0))

plt.savefig('feature_stat_swarmplot_' + stringency + '.svg', bbox_inches = 'tight')
```
