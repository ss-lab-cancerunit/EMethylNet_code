# Index to probe name

Converts a feature importances file which contains probe indices to the corresponding list of probe names

```python
cancer_type = ''
i = 0 # currently build for looking at one run, but would be easy to change to look at multiple
```

```python
# read in the whole probe list
import pandas as pd
if cancer_type == '':
    probes = pd.read_csv('../../data_preprocessing/dataset/pandas/m_values/TCGA-all.csv', sep = '\t', usecols=[0])
else:
    probes = pd.read_csv('../../data_preprocessing/dataset/pandas/m_values/TCGA-'+cancer_type+'.csv', sep = '\t', usecols=[0])

```

```python
# read in the feature_importances
features = pd.read_csv(cancer_type + '_run_' + str(i) + '.csv', header = None)
feature_indices = [int(features[0][i].split(' ')[2]) for i in range(len(features[0]))]
```

```python
wanted_probes = probes['probe'][feature_indices]

# could write to file here, for now I just print them
for i in wanted_probes:
    print(i)
```
